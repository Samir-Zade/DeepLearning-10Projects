# **Isolation Forest Anomoly Detection**

```python
#import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
#setting the seed value
random_state = 47

# Generate shuffled arrays for x and y
x_values = np.linspace(-6, 10, 350)
y_values = np.linspace(-6, 10, 350)
np.random.shuffle(x_values)
np.random.shuffle(y_values)

# Create DataFrame
df = pd.DataFrame({'0': x_values, '1': y_values})
df.head()
```

```python
#create the plot of data points
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
```

```python
#import the isolation forest
from sklearn.ensemble import IsolationForest

#create the model
clf = IsolationForest(contamination = "auto")

#fit and predict the model
clf.fit(df)
prediction = clf.predict(df)
```

```python
prediction
```

```python
#find the outliers which are below 0
index = np.where(prediction < 0)
index
```

```python
x = df.values
x
```

```python
#plot the scatter plot
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.scatter(x[index, 0], x[index, 1], edgecolors = 'r')
plt.show()
```
