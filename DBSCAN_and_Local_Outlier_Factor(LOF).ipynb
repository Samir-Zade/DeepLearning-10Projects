# **DBSCAN Clustering**

```python
#import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
```

```python
#create the data
X, y = make_circles(n_samples = 750, factor = 0.3, noise = 0.1)
X
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

```python
#create the dbscan model
dbscan = DBSCAN(eps = 0.15)

#fit and predict
dbscan.fit_predict(X)
```

```python
#find the labels_
dbscan.labels_
```

```python
plt.scatter(X[:, 0], X[:, 1], c = dbscan.labels_)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = y)
```

# **Local outlier anomoly detection**

```python
# Create DataFrame
df = pd.DataFrame({'0': X[:, 0], '1': X[:, 1]})
df.head()
```

```python
#import the libary
from sklearn.neighbors import LocalOutlierFactor

#create the model
lof = LocalOutlierFactor(contamination = "auto", n_neighbors = 5)

#fit and predict
pred = lof.fit_predict(df)
```

```python
#find the outliers which are below 0
index = np.where(pred < 0)
index
```

```python
x = df.values
x
```

```python
#plot the scatter plot
plt.scatter(df.iloc[:, 0], df.iloc[:, 1])
plt.scatter(x[index, 0], x[index, 1], edgecolors = 'r')
plt.show()
```
