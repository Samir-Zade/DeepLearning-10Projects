```python
#importing the required libarires
import numpy as np
import pandas as pd
```

```python
print(1e-4)
```

```python
#create class for avoid repetation of code

class Perceptron:
  def __init__(self, eta, epochs):
    #weights = 3 because the perceptron model is designed to work with input data that has three features
    self.weights = np.random.randn(3) * 1e-4
    print(f"self.weights : {self.weights}")

    self.eta = eta #eta is learning rate
    self.epochs = epochs #epochs is iteration of weight updations


  def activationFunction(self, inputs, weights):
    z = np.dot(inputs, weights)
    return np.where(z > 0, 1, 0)

  def fit(self, X, y):
    self.X = X
    self.y = y


    #Including a bias term allows the perceptron to capture patterns that don't necessarily pass through the origin (0,0) in the input space.
    X_with_bias = np.c_[self.X, -np.ones((len(self.X), 1))]
    print(f"X_with_bias : {X_with_bias}")

    for epoch in range(self.epochs):
      print(f"For epoch : {epoch}")
      y_hat = self.activationFunction(X_with_bias, self.weights)
      print(f"predicted values : \n{y_hat}")

      error = self.y - y_hat
      print(f"error : \n{error}")

      #weight updation formula :
      self.weights = self.weights + self.eta * np.dot(X_with_bias.T, error)
      print(f"Updated weights : \n{self.weights}")

  def predict(self, X):
      X_with_bias = np.c_[X, -np.ones((len(self.X), 1))]
      return self.activationFunction(X_with_bias, self.weights)
```

## Using AND GATE

```python
#create the data
data = {"x1" : [0, 0, 1, 1], "x2" : [0, 1, 0, 1], "y" : [0, 0, 0, 1]}

AND = pd.DataFrame(data)
AND
```

```python
X = AND.drop("y", axis = 1)
X
```

```python
y = AND["y"]
y.to_frame()
```

```python
model = Perceptron(eta = 0.5, epochs = 10)
model.fit(X, y)
```

```python
model.predict(X)
```

```python
model.weights
```

## Using XOR GATE
- it is complex gate type, so perceptron are not familiar with complex relationship of models.

```python
#create the data
data = {"x1" : [0, 0, 1, 1], "x2" : [0, 1, 0, 1], "y" : [0, 1, 1, 0]}

XOR = pd.DataFrame(data)
XOR
```

```python
X = XOR.drop("y", axis = 1)
X
```

```python
y = XOR.y
y.to_frame()
```

```python
model = Perceptron(0.5, 100)
model.fit(X, y)
```

```python
model.predict(X)
```

Here, we can find that the model is not accessable for finding the complex relationship between the datasets

- predicted output = array([0, 0, 0, 0])
- actual output required = array([0, 1, 1, 0])

But, we can find them by using some changes.
